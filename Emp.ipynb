```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('/content/Extended_Employee_Performance_and_Productivity_Data.csv')
```

```python
# Check the first few rows of the dataset
print("First 5 rows of the dataset:")
print(data.head())
```

```python
# Get information about the dataset (column names, data types, etc.)
print("\nDataset Info:")
print(data.info())
```

```python
# Get summary statistics of the numerical columns
print("\nSummary Statistics:")
print(data.describe())
```

```python
 sns.histplot(data['Monthly_Salary'], kde=True, bins=30, color='yellow' )

# Add title and labels
plt.title('Histogram of Employee Salaries ', fontsize=15)
plt.xlabel(' value')
plt.ylabel('Frequency', fontsize=12)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each job title
job_title_counts = data['Job_Title'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(job_title_counts, labels=job_title_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

# Add title
plt.title('Distribution of Employees by Job Title', fontsize=15)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each gender
gender_counts = data['Gender'].value_counts()

# Plot the bar graph
plt.figure(figsize=(8, 6))  # Set the figure size
gender_counts.plot(kind='bar', color='lightcoral', edgecolor='black')

# Add title and labels
plt.title('Distribution of Employees by Gender', fontsize=15)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)

# Show the plot
plt.xticks(rotation=0)  # Keep x-axis labels horizontal for better readability
plt.show()
```

```python
# Calculate mean, median, and mode for the 'MonthlyIncome' column
mean_income = data['Monthly_Salary'].mean()
median_income = data['Monthly_Salary'].median()
mode_income = data['Monthly_Salary'].mode()[0]

# Print the results
print(f"Mean Monthly Income: {mean_income}")
print(f"Median Monthly Income: {median_income}")
print(f"Mode Monthly Income: {mode_income}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(x=data['Job_Title'], color='lightblue')

plt.title('Box Plot of Monthly Income', fontsize=15)
plt.xlabel(' Job title (Job_Title)', fontsize=12)

plt.show()
```

```python
import numpy as np

df = data['Work_Hours_Per_Week']

variance = np.var(df, ddof=1)
std_dev = np.std(df, ddof=1)
range_value = np.max(df) - np.min(df)
iqr = np.percentile(df, 75) - np.percentile(df, 25)

print(variance, std_dev, range_value, iqr)
```
