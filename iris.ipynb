```python
species_list = ["Iris-setosa", "Iris-versicolor", "Iris-virginica"]
filtered_df = data[data["species"].isin(species_list)]
for species in species_list:
    print(f"Statistics for {species}:")
    species_df = filtered_df[filtered_df["species"] == species].drop(columns=["species"])
    print(species_df.describe(percentiles=[.25, .5, .75]))
    print("\n")
```
