```python


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the dataset
file_path = "/content/Titanic-Dataset.csv"  # Ensure this matches the uploaded file name
titanic = pd.read_csv(file_path)

# Convert column names to lowercase for consistency
titanic.columns = titanic.columns.str.lower()

# Display first few rows to check dataset structure
print("First 5 rows of the dataset:")
print(titanic.head())

# Display column names
print("\nColumn Names in Dataset:")
print(titanic.columns)

# Step 2: Check for missing values
print("\nMissing Values in Dataset:")
print(titanic.isnull().sum())

# -----  Pattern Analysis with Seaborn -----

# 1 Correlation Heatmap - Checking Relationships Between Features
plt.figure(figsize=(8, 6))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title(" Correlation Heatmap - Titanic Dataset")
plt.show()

# 2 Boxplot: Checking Fare Distribution by Passenger Class
plt.figure(figsize=(8, 6))
sns.boxplot(x='pclass', y='fare', data=titanic, palette='coolwarm')
plt.xlabel("Passenger Class")
plt.ylabel("Ticket Fare")
plt.title(" Fare Distribution Across Passenger Classes")
plt.show()

# 3 Countplot: Checking Survival Rate by Gender
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', hue='survived', data=titanic, palette='Set2')
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title(" Survival Rate by Gender")
plt.legend(["Did Not Survive", "Survived"])
plt.show()

# -----  Histogram of Ticket Prices -----
plt.figure(figsize=(10, 6))
sns.histplot(titanic['fare'], bins=30, kde=True, color='blue')

# Labels and title
plt.xlabel('Ticket Fare')
plt.ylabel('Number of Passengers')
plt.title(' Distribution of Ticket Prices on the Titanic')

# Show the plot
plt.show()
```

```python
# Step 1: Upload the CSV file in Google Colab
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Step 2: Load the Titanic dataset from the uploaded CSV file
file_path = "/content/Titanic-Dataset.csv"  # Make sure the file name matches exactly
titanic = pd.read_csv(file_path)

# Convert column names to lowercase for consistency
titanic.columns = titanic.columns.str.lower()

# Plot box plot for age distribution grouped by gender and survival status
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='age', hue='survived', data=titanic, palette='Set2')

# Labels and title
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title(' Age Distribution by Gender and Survival Status')

# Show the plot
plt.show()

# ---- Print Observations Based on the Box Plot ----
print("\n Observations from the Box Plot:\n")
print("1 Younger passengers had higher survival rates.")
print("   - The median age of survivors (especially for females) is lower than non-survivors.")
print("   - Indicates that younger passengers had better chances of survival.")

print("\n2 Women had a higher survival rate.")
print("   - The box for survived females is higher compared to males.")
print("   - Few young men survived, reinforcing the 'women and children first' rule.")

print("\n3 Men had lower survival rates across all ages.")
print("   - The median age of surviving males is lower than non-survivors.")
print("   - Older men had a much lower chance of survival.")

print("\n4 Outliers present in older age groups.")
print("   - Some elderly passengers survived, but they were rare exceptions.")

print("\n Conclusion: The data supports the 'Women and children first' evacuation rule.")
```
